```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df_r5 = pd.read_csv('r5.csv')

# Replace this with the actual section names you are interested in
selected_sections = ['()', '', '']

# Convert 'start_time' to datetime, extract the hour, and sort
df_r5['hour'] = pd.to_datetime(df_r5['start_time']).dt.hour
df_r5['percentage'] = df_r5['percentage'].astype(float)  # Ensure 'percentage' is a float

# Calculate the average percentage for each hour of the day for each section
hourly_avg = df_r5.groupby(['section', 'hour'])['percentage'].mean().reset_index()

# Plot each section in a separate plot
for section in selected_sections:
    section_hourly_avg = hourly_avg[hourly_avg['section'] == section]

    # Create a bar plot for the section
    plt.figure(figsize=(12, 6))
    barplot = sns.barplot(data=section_hourly_avg, x='hour', y='percentage', color='blue')

    # Set the title, x-axis, and y-axis labels
    plt.title(f'Average Usage Percentage by Hour')
    plt.xlabel('Hour of the Day')
    plt.ylabel('Average Percentage In Use')

    # Show the plot
    plt.tight_layout()
    plt.show()
```
